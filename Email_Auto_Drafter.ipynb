### Problem Statement  
To design and implement an AI-based Email Auto-Drafter that converts bullet-point inputs into a polite and professional email using Hugging Face text generation pipelines.


### Objective
The objective of this project is to demonstrate the use of Generative AI models for real-world productivity tasks.  
This project focuses on using Hugging Face pipelines to automatically generate professional emails from minimal user input.


### Technology Used
- Python  
- Hugging Face Transformers  
- Pretrained Model: `distilgpt2`  
- Task: Text Generation


```python
# Install required library (run once)
!pip install transformers
```

```python
# Import required module
from transformers import pipeline
```

```python
# Initialize the text-generation pipeline
email_generator = pipeline(
    "text-generation",
    model="distilgpt2"
)
```

### Sample Input
The user provides bullet-point information instead of writing a full email.

Example Input:
- Sick leave  
- Monday  
- Back on Tuesday  


```python
# Prompt Engineering
prompt = """
Write a polite and professional email using the following points:
- Sick leave
- Monday
- Back on Tuesday
"""
```

```python
output = email_generator(
    prompt,
    max_new_tokens=120,   # Correct parameter
    do_sample=True,
    temperature=0.7,
    truncation=True
)

print(output[0]["generated_text"])
```

### Sample Output
The model generates a formal email based on the bullet points provided by the user.

Example Output:
> Dear Sir/Madam,  
> I am writing to inform you that I will be on sick leave this Monday and will resume work on Tuesday.  
> Thank you for your understanding.
